```python
### 小試身手-2

# 1. 計算 1+2+3+…+100 之總和

# 2. 試撰寫一程式 , 列印出以下圖形
#          ******
#          *****
#          ****
#          ***
#          **
#          *


# 3.  有一變數 name=“Jack!” , 試撰寫一程式 , 列印出下列訊息
#         JJJJJ
#         aaaa
#         ccc
#         kk
#         !

# 4. 有一 list = [“a”, “b”, “c”, “c”, “c”, “a”, “d”, “b”, “b”, “a”, “c”]
#     試統計各字母出現次數 , 並以 dict 型式表達

### word count 

## 保留字
data_list = ["a","b","c","c","c","a","d","b","b","a","c","c","c","c","e","f","k","a","f"]

# {
#     "a" : 3,
#     "b" : 5,
#     "c" : 1,
#     "d" : 2
# }

result = {}

for ele in data_list:
    # print(ele)
    if ele in result:
        result[ele] += 1
    else:
        result[ele] = 1
    
print(result)


```
