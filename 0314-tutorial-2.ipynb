```python
### func-2

def get_scores_info(scores,name):
    # print(scores)
    print("{} Sum :".format(name), sum(scores))
    print("{} Avg :".format(name), sum(scores)/len(scores))
    print("{} Max :".format(name), sorted(scores,reverse=True)[0]  )



math    = [80,97,95,87,51]
eng     = [100,68,78,90,100]
chinese = [75,50,65,62,88]

get_scores_info(math,"Math")

print("-"*80)

get_scores_info(eng,"English")

print("-"*80)

get_scores_info(chinese,"Chinese")

```
